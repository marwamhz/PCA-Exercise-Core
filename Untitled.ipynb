```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

from sklearn import set_config
set_config(transform_output='pandas')
pd.set_option('display.max_columns',100)
```

## 1. Load the Data

```python
# Load the Data
fpath = "Data/NHANES_Diet_Diabetes_Risk.xls.csv"
df = pd.read_csv(fpath)
df.info()
df.head()
```

## 2. Preprocess the Data

```python
# Set SEQN as the index
df = df.set_index("SEQN")
```

```python
# define X and y
X = df.drop(columns = ['Diabetes_Risk'])
y = df['Diabetes_Risk']
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# check missing values 
import missingno as msno
msno.matrix(X)
```

```python
# Impute Missing Values with the median value
## Instantiate SimpleImputer
impute_median = SimpleImputer(strategy='median')
# Fit & transform data.
X_train_tf = impute_median.fit_transform(X_train)
X_test_tf = impute_median.transform(X_test)
```

```python
# Scale
## Instantiate Standard Scaler
scaler = StandardScaler()
## Fit & transform data.
X_train_scaled = scaler.fit_transform(X_train_tf)
X_test_scaled = scaler.transform(X_test_tf)
```

## 3. Model - No PCA

```python
# Define Custom Functions
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

def classification_metrics(y_true, y_pred, label='',
                           output_dict=False, figsize=(8,4),
                           normalize='true', cmap='Blues',
                           colorbar=False):
    # Get the classification report
    report = classification_report(y_true, y_pred)
    ## Print header and report
    header = "-"*70
    print(header, f" Classification Metrics: {label}", header, sep='\n')
    print(report)
    ## CONFUSION MATRICES SUBPLOTS
    fig, axes = plt.subplots(ncols=2, figsize=figsize)
    # create a confusion matrix  of raw counts
    ConfusionMatrixDisplay.from_predictions(y_true, y_pred,
                normalize=None, cmap='gist_gray', colorbar=colorbar,
                ax = axes[0],);
    axes[0].set_title("Raw Counts")
    # create a confusion matrix with the test data
    ConfusionMatrixDisplay.from_predictions(y_true, y_pred,
                normalize=normalize, cmap=cmap, colorbar=colorbar,
                ax = axes[1]);
    axes[1].set_title("Normalized Confusion Matrix")
    # Adjust layout and show figure
    fig.tight_layout()
    plt.show()
    # Return dictionary of classification_report
    if output_dict==True:
        report_dict = classification_report(y_true, y_pred, output_dict=True)
        return report_dict


def evaluate_classification(model, X_train, y_train, X_test, y_test,
                         figsize=(6,4), normalize='true', output_dict = False,
                            cmap_train='Blues', cmap_test="Reds",colorbar=False):
    # Get predictions for training data
    y_train_pred = model.predict(X_train)
    # Call the helper function to obtain regression metrics for training data
    results_train = classification_metrics(y_train, y_train_pred, #verbose = verbose,
                                     output_dict=True, figsize=figsize,
                                         colorbar=colorbar, cmap=cmap_train,
                                     label='Training Data')
    print()
    # Get predictions for test data
    y_test_pred = model.predict(X_test)
    # Call the helper function to obtain regression metrics for test data
    results_test = classification_metrics(y_test, y_test_pred, #verbose = verbose,
                                  output_dict=True,figsize=figsize,
                                         colorbar=colorbar, cmap=cmap_test,
                                    label='Test Data' )
    if output_dict == True:
        # Store results in a dataframe if ouput_frame is True
        results_dict = {'train':results_train,
                    'test': results_test}
        return results_dict

```

```python
from sklearn.neighbors import KNeighborsClassifier
# Instantiate default KNN
knn = KNeighborsClassifier()
# Fit a default KNN model to the data.
knn.fit(X_train_scaled, y_train)
```

```python
%%time
# Record the time it takes to make predictions
preds = knn.predict(X_test_scaled)
```

```python
# Obtain evaluation metrics using custom function
evaluate_classification(knn, X_train_scaled,y_train, X_test_scaled, y_test)
```

## 5. Model - With PCA

```python
# Instantiate & fit data using PCA
pca = PCA(n_components=3)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
# Fit a default KNN model to the PCA data.
knn.fit(X_train_pca, y_train)
```

```python
%%time
# Record the time it takes to make predictions
preds_pca = knn.predict(X_test_pca)
```

```python
# Obtain evaluation metrics using custom function
evaluate_classification(knn, X_train_pca,y_train, X_test_pca, y_test)
```
